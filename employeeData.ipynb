```python
import matplotlib.pyplot as plt
# 1. Diagram Batang
age_stats = df.groupby('Departemen')['Usia'].agg(['max', 'min'])

plt.figure(figsize=(12, 6))
age_stats.plot(kind='bar', width=0.8)

plt.title('Statistik Usia per Departemen')
plt.xlabel('Departemen')
plt.ylabel('Usia')
plt.legend(['Usia Maksimum', 'Usia Minimum', 'Selisih Usia'])
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 2. Diagram Lingkar
plt.figure(figsize=(4, 4))

df['TingkatPendidikan'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Proporsi Karyawan berdasarkan Tingkat Pendidikan')
plt.ylabel('20 Karyawan')

plt.tight_layout()
plt.show()
```

```python
# 2. Diagram Lingkar
plt.figure(figsize=(4, 4))

df['JenisKelamin'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Proporsi Karyawan berdasarkan Jenis Kelamin')
plt.ylabel('20 Karyawan')

plt.tight_layout()
plt.show()
```

```python
# 3. Diagram Garis
plt.figure(figsize=(8, 4))

df.groupby('LamaBekerja')['GajiTahunan'].mean().plot(kind='line', marker='o')
plt.title('Gaji Tahunan berdasarkan Lama Bekerja')
plt.xlabel('Lama Bekerja (tahun)')
plt.ylabel('Rata-rata Gaji Tahunan')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
# 4. Histogram
plt.figure(figsize=(12, 6))

sns.histplot(df['LamaBekerja'], bins=10, kde=True)
plt.title('Distribusi Lama Bekerja')
plt.xlabel('Lama Bekerja (Tahun)')
plt.ylabel('Jumlah Karyawan')
plt.tight_layout()

plt.show()
```

```python
# 5. Scatter Plot
plt.figure(figsize=(12, 6))

sns.scatterplot(x='JumlahProyek', y='NilaiKinerja', data=df)
plt.title('Hubungan antara Jumlah Proyek dan Nilai Kinerja')
plt.xlabel('Jumlah Proyek')
plt.ylabel('Nilai Kinerja')

plt.tight_layout()
plt.show()
```

```python
# 6. Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Departemen', y='TingkatKepuasan', data=df)
plt.title('Distribusi Tingkat Kepuasan per Departemen')
plt.xlabel('Departemen')
plt.ylabel('TingkatKepuasan')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
